Label Encoding
 . converting the labels into numeric form

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
#loading the data from csv file to pandas dataframe
cancer_data = pd.read_csv('/data.csv')
cancer_data.head()
```

```python
#finding the count of different labels
cancer_data['diagnosis'].value_counts()
```

```python
#load the label encoder function
label_encode = LabelEncoder()
```

```python
label = label_encode.fit_transform(cancer_data.diagnosis)
```

```python
#appending the label to the dataframe
cancer_data['target'] = label
```

```python
cancer_data.head()
```

 - 0 ---> Benign
 - 1 ---> Malignant

```python
cancer_data['target'].value_counts()
#
```

Label Encoding of Iris Data

```python
iris_data = pd.read_csv('/iris_data.csv')
iris_data.head()
```

```python
iris_data['Species'].value_counts()
```

```python
#load label encoder
label_encoder_1 = LabelEncoder()
```

```python
iris_labels = label_encoder_1.fit_transform(iris_data.Species)
```

```python
iris_data['target'] = iris_labels
```

```python
iris_data.head()
```

```python
iris_data['target'].value_counts()
```

- Iris-Setosa --> 0
- Iris-Versicolor --> 1
- Iris-Verginica --> 2
