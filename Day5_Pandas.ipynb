Manipulating a DataFrame

```python
#adding a column to a dataframe
housing_df['MedInc_doubled'] = housing_df['MedInc']*2
housing_df.head()
```

```python
# removing a row
housing_df.drop(index=0, axis=0)
```

```python
# drop a column
housing_df.drop(columns='MedInc_doubled', axis=1)
```

```python
#locating a row using the index value
housing_df.iloc[1]
```

```python
#locating a particular column
print(housing_df.iloc[:,0]) #first column
print(housing_df.iloc[:,-1]) #last column
print(housing_df.iloc[:,0:2]) #first two
```
